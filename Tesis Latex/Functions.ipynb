```python

# plot inputs and outputs
from matplotlib import pyplot
import matplotlib.patches as mpatches
import numpy as np
```

```python
# rectified linear function
def rectified(x):
	return max(0.0, x)
def rectified_prime(x):
    if(x<0):
        return 0
    else:
        return 1

    
series_in = [x for x in np.arange(-1.5, 1.51,.01)]
series_out = [rectified(x) for x in series_in]

series_out_p = [rectified_prime(x) for x in series_in]
# line plot of raw inputs to rectified outputs
fig0 = pyplot.figure(0,figsize=(8,5))
pyplot.rcParams["font.family"] = "times new roman"
pyplot.xticks(size = 17) 
pyplot.yticks(size = 17)
pyplot.plot(series_in, series_out,linewidth=2.5)
pyplot.plot(series_in, series_out_p,linewidth=2.5)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g (x)')
or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función ReLU",fontsize=20)
pyplot.rc('legend', fontsize=16) 
pyplot.axhline(0, color='gray')
pyplot.axvline(0, color='gray')
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('ReLU.png')
pyplot.show()
```

```python
# Sigmoid
x = np.linspace(-5, 5, 200) 
z = 1/(1 + np.exp(-x)) 
z_prime = z*(1-z)

fig0 = pyplot.figure(0,figsize=(8,5))
pyplot.rcParams["font.family"] = "times new roman"
pyplot.xticks(size = 17) 
pyplot.yticks(size = 17)
pyplot.plot(x, z,linewidth=2.5)
pyplot.plot(x, z_prime,linewidth=2.5)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g (x)')
or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función Sigmoide",fontsize=20)
pyplot.rc('legend', fontsize=16) 
pyplot.axhline(0, color='gray')
pyplot.axvline(0, color='gray')
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('Sigmoid.png')
pyplot.show()
```

```python
# TanH
x = np.linspace(-5, 5, 200) 
z = np.tanh(x) 
z_prime = 1 - z**2

fig0 = pyplot.figure(0,figsize=(8,5))
pyplot.rcParams["font.family"] = "times new roman"
pyplot.xticks(size = 17) 
pyplot.yticks(size = 17)
pyplot.plot(x, z,linewidth=2.5)
pyplot.plot(x, z_prime,linewidth=2.5)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g (x)')
or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función TanH", fontsize=20)
pyplot.rc('legend', fontsize=16) 
pyplot.axhline(0, color='gray')
pyplot.axvline(0, color='gray')
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('TanH.png')
pyplot.show()
```

```python
# rectified linear function
def Lrectified(x,a):
	return max(x*a, x)
def u(x):
    if(x<0):
        return 0
    else:
        return 1
    
def Lrectified_prime(x,a):
    return a + (1-a)*u(x)
 
# define a series of inputs
series_in = [x for x in np.arange(-1.5, 1.51,.01)]
series_out = [Lrectified(x,.2) for x in series_in]

series_out_p = [Lrectified_prime(x,.2) for x in series_in]
# line plot of raw inputs to rectified outputs
fig0 = pyplot.figure(0,figsize=(8,5))
pyplot.rcParams["font.family"] = "times new roman"
pyplot.xticks(size = 17) 
pyplot.yticks(size = 17)
pyplot.plot(series_in, series_out,linewidth=2.5)
pyplot.plot(series_in, series_out_p,linewidth=2.5)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g (x)')
or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función PReLU", fontsize=20)
pyplot.rc('legend', fontsize=16) 
pyplot.axhline(0, color='gray')
pyplot.axvline(0, color='gray')
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('LeakyReLU.png')
pyplot.show()
```

```python
# rectified linear function
def elu(arr):
    a = []
    for x in arr:
        if x >= 0:
            a.append(x)
        else:
            a.append((np.exp(x)-1))
    return a
    
def elu_prime(arr):
    a = []
    for x in arr:
        if x >= 0:
            a.append(1)
        else:
            a.append(np.exp(x))
    return a
 
# define a series of inputs
series_in = [x for x in np.arange(-3.51, 3.01,.01)]
series_out = elu(series_in)

series_out_p = elu_prime(series_in)
# line plot of raw inputs to rectified outputs
fig0 = pyplot.figure(0,figsize=(8,5))
pyplot.rcParams["font.family"] = "times new roman"
pyplot.xticks(size = 17) 
pyplot.yticks(size = 17)
pyplot.plot(series_in, series_out,linewidth=2.5)
pyplot.plot(series_in, series_out_p,linewidth=2.5)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g (x)')
or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función ELU", fontsize=20)
pyplot.rc('legend', fontsize=16) 
pyplot.axhline(0, color='gray')
pyplot.axvline(0, color='gray')
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('ELU.png')
pyplot.show()
```
