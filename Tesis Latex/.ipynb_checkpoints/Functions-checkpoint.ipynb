```python

# plot inputs and outputs
from matplotlib import pyplot
import matplotlib.patches as mpatches
import numpy as np
```

```python
# rectified linear function
def rectified(x):
	return max(0.0, x)
def rectified_prime(x):
    if(x<0):
        return 0
    else:
        return 1
 
# define a series of inputs
series_in = [x for x in np.arange(-1.5, 1.51,.01)]
series_out = [rectified(x) for x in series_in]

series_out_p = [rectified_prime(x) for x in series_in]
# line plot of raw inputs to rectified outputs
pyplot.plot(series_in, series_out)
pyplot.plot(series_in, series_out_p)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g(x)')
or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función ReLU")
pyplot.rc('legend', fontsize=14) 
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('ReLU.png')
pyplot.show()
```

```python
# rectified linear function
x = np.linspace(-5, 5, 200) 
z = 1/(1 + np.exp(-x)) 
z_prime = z*(1-z)

# line plot of raw inputs to rectified outputs
pyplot.plot(x, z)
pyplot.plot(x, z_prime)
pyplot.grid(True) 
blue_patch = mpatches.Patch(color='blue', label='g(x)')
#or_patch = mpatches.Patch(color='orange', label='g\'(x)')
pyplot.title("Función Sigmoide")
pyplot.rc('legend', fontsize=14) 
pyplot.legend(handles=[blue_patch,or_patch])
pyplot.savefig('ReLU.png')
pyplot.show()
```
